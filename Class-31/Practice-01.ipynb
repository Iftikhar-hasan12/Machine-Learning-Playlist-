```python
data = pd.read_csv("diabetes.csv")
```

```python
# Take user input
user_input = input("Enter values for 8 features (comma-separated):\n"
                   "Pregnancies, Glucose, BloodPressure, SkinThickness, Insulin, BMI, DiabetesPedigreeFunction, Age:\n")

# Convert input to numpy array
user_values = np.array([float(x) for x in user_input.strip().split(',')]).reshape(1, -1)

# Predict
user_pred = model.predict(user_values)
prob = model.predict_proba(user_values)[0][1]

# Show result
if user_pred[0] == 1:
    print(f"ðŸ”´ Prediction: you have diabetes (Probability: {prob:.2f})")
else:
    print(f"ðŸŸ¢ Prediction: You have NOT have diabetes (Probability: {prob:.2f})")
```
