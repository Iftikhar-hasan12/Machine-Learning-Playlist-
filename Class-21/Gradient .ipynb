```python
import numpy as np
import matplotlib.pyplot as plt

# Step 1: Generate sample data
np.random.seed(0)
X = 2 * np.random.rand(100, 1)
y = 4 + 3 * X + np.random.randn(100, 1)

# Flatten X and y for simplicity
X_flat = X.flatten()
y_flat = y.flatten()

# Step 2: Implement Gradient Descent
def gradient_descent(X, y, lr, n_iterations ):
    m = 0
    b = 0
    N = len(y)

    for i in range(n_iterations):
        y_pred = m * X + b
        dm = (-2/N) * np.sum(X * (y - y_pred))
        db = (-2/N) * np.sum(y - y_pred)
        m = m - lr * dm
        b = b - lr * db

    return m, b

# Run gradient descent
n_iterations =100
lr=0.01
m_final, b_final = gradient_descent(X_flat, y_flat,lr,n_iterations )

# Predicted y using gradient descent
y_pred_gd = m_final * X_flat + b_final

# Step 3: Plot
plt.figure(figsize=(8, 5))
plt.scatter(X, y, color='blue', alpha=0.6, label='Data Points')
plt.plot(X, y_pred_gd, color='orange', linewidth=2, label='Gradient Descent Line')
plt.title("Best Fit Line using Gradient Descent")
plt.xlabel("X")
plt.ylabel("y")
plt.legend()
plt.grid(True)
plt.show()
```

```python
def compute_cost(X, y, m, b):
    y_pred = m * X + b
    return np.mean((y - y_pred) ** 2)
```

```python
def gradient_descent(X, y, lr=0.01, n_iterations=20):
    m = 0
    b = 0
    N = len(y)
    
    history = []

    for i in range(n_iterations):
        y_pred = m * X + b
        dm = (-2/N) * np.sum(X * (y - y_pred))
        db = (-2/N) * np.sum(y - y_pred)
        m = m - lr * dm
        b = b - lr * db
        cost = compute_cost(X, y, m, b)
        history.append((m, b, cost))

    return m, b, history
```
